#소수 구하기
    에라토스테네스의 체의 활용 유무에 따른 동작시간 측정
    - case 1 : 에라토스 테네스의 체
    - case 2 : 직관적으로 작성한 코드

    샘플의 수가 적은 경우 중 몇번을 제외하고 case 1이 항상 빠르다. 
    이는 case 1의 시간복잡도가 N*logN으로 선형에 가깝기 때문에, 
    N**2의 시간복잡도를 가진 case2보다 빠르다.
    
    다만 메모리를 미리 많이 할당해 둔다는 단점이 있다.

```python
import math
import timeit
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#case1
def case1(n):
    start = timeit.default_timer()
    array = [True] * (n+1)

    for i in range(2, int(math.sqrt(n)+1)):
        if array[i] == True:
            j = 2
            while i*j <= n:
                array[i*j] = False
                j+=1
    #for i in range(2, n+1):
        #if array[i]:
            #print(i,end = " ")
    stop = timeit.default_timer()
    time = stop - start
    #time = round(stop - start,6)
    #print(time)
    return time
#print()
```

```python
#case2
#%%time
def case2(n):
    start = timeit.default_timer()
    #Prime = []
    count=0
    a = 1
    b = n
    for i in range(a,b+1):
        if i == 2 or i == 3:
            count+=1
            #Prime.append(i)
            continue
        for j in range(2,i//2+2):
            if i%j==0:
                break
            if j == i//2+1:
                count+=1
                #Prime.append(i)
    #print("소수개수 {}".format(count))
    #print(Prime)
    stop = timeit.default_timer()
    time = stop - start
    return time
```

1~40

```python
#plt.rcParams['figure.figsize'] = (10,6)
x = [x for x in range(40)]
y = [case1(x) for x in x]
z = [case2(x) for x in x]
plt.xlabel("Data numbers(unit : ea)")
plt.ylabel("time")
plt.plot(x,y, linestyle = "-", label = "case1")
plt.plot(x,z, linestyle = "-", label = "case2")
plt.legend()
plt.show()
```

100 ~ 10000

```python
x = [x*100 for x in range(1,100)]
y = [case1(x) for x in x]
z = [case2(x) for x in x]

plt.xlabel("Data numbers(unit : ea)")
plt.ylabel("time(unit : s)")
plt.plot(x,y, linestyle = "-", label = "case1")
plt.plot(x,z, linestyle = "-", label = "case2")
plt.legend()
plt.show()
```
