큐 - First In, First Out

    isEmpty - 큐가 비었는지 확인하는 함수
    enqueue - 큐에 삽입하는 함수
    dequeue - 큐에서 제거하는 함수
    peek - 큐가 가진 것중 가장 먼저 삽입된 데이터를 확인하는 함수
    size - 큐가 가진 데이터의 개수를 확인하는 함수
    clear - 큐를 초기화 하는 함수

```python
#큐
class queue:
    def __init__(self):
        self.q = []
    def queue(self):
        self.q=[]
    def isEmpty(self):
        if len(self.q) == 0:
            print("Queue is empty")
            return True
        else:
            return False
    def enqueue(self,x):
        self.q.append(x)
        print(f"{x} is inserted")

    def dequeue(self):
        val = self.q.pop(0)
        return val

    def peek(self):
        return self.q[0]

    def size(self):
        return len(self.q)

    def clear(self):
        self.__init__()
        print("queue is empty")
    def get_queue(self):
        print(self.q)

def pathfinder4(i, j, y, x):
    nq1=queue()
    nq1.enqueue([i,j])
    while 1:
        map[i][j]=2
        print("move to ({}, {})".format(i,j))
        if y==i and x==j:
            print("Finished!")
            break
        if i+1<map_size and map[i+1][j]==0:
            nq1.enqueue([i+1,j])
        if i>0 and map[i-1][j]==0:
            nq1.enqueue([i-1,j])
        if j+1<map_size and map[i][j+1]==0:
            nq1.enqueue([i,j+1])
        if j>0 and map[i][j-1]==0:
            nq1.enqueue([i,j-1])
        #중복 제거 옵션
        while 1:
            [i,j] = nq1.dequeue()
            if map[i][j] == 0:
                break

    for i in range(map_size):
        print(map[i])
```

원형 큐

     - 기본적인 함수는 큐와 동일하지만, 기본 큐와 다르게 크기가 있기때문에, 전단과 후단이 겹치지 않도록 하는 장치가 있다.
       갈림길이 많은 경우 삽입과정에서 데이터 손실이 발생할 수 있다. 
       크기가 정해져 있기때문에 전단의 데이터를 삭제하더라도 값의 주소값을 이동하지 않기때문에 더 빠른 경우가 있다.

```python
#원형 큐
class n_queue:
    def __init__(self):
        self.max_size=int(input())
        self.front=0
        self.rear=0
        self.q=[]
    def queue(self):
        self.q=[None]*self.max_size
    def isEmpty(self):
        if self.front==self.rear:
            return True
        else:
            return False
    def enqueue(self,x):
        if self.front%self.max_size == (self.rear+1)%self.max_size:
            print("queue is full.")
        else:
            self.q[self.rear]=x
            self.rear=(self.rear+1)%self.max_size
            print("{} is inserted".format(x))
    def dequeue(self):
        val=self.q[self.front]
        self.q[self.front]=None
        self.front=(self.front+1)%self.max_size
        return val
    def peek(self):
        return self.q[self.front]
    def size(self):
            return (self.rear+self.max_size-self.front)%self.max_size
    def clear(self):
        self.__init__()
        self.queue()
        print("queue is empty")
    def get_queue(self):
        print(self.q)
```

```python
def pathfinder2(i, j, y, x):
    nq=n_queue()
    nq.queue()
    nq.enqueue([i,j])
    while 1:
        map[i][j]=2
        print("move to ({}, {})".format(i,j))
        if y==i and x==j:
            print("Finished!")
            break
        if i+1<map_size and map[i+1][j]==0:
            nq.enqueue([i+1,j])
        if i>0 and map[i-1][j]==0:
            nq.enqueue([i-1,j])
        if j+1<map_size and map[i][j+1]==0:
            nq.enqueue([i,j+1])
        if j>0 and map[i][j-1]==0:
            nq.enqueue([i,j-1])
        while 1:
            [i,j] = nq.dequeue()
            if map[i][j] == 0:
                break

    for i in range(map_size):
        print(map[i])
```

우선순위 큐

    - 기본적인 기능은 큐와 동일하지만, 거리값을 통해 우선순위를 판별할 수 있다.
      본 클래스에서는 거리값 * (-1)를 하고, max()를 통해 가장 거리가 가까운 값을 찾아서 사용한다.

```python
class Priority_Queue:
    def __init_(self):
        self.pq=[]
    def PriorityQueue(self):
        self.pq=[]
    def isEmpty(self):
        if len(self.pq)==0:
            return True
        else:
            return False
    def enqueue(self,e):
        self.pq.append(e)
        print('{} inserted'.format(e))
    def dequeue(self):
        return self.pq.pop(self.pq.index(max(self.pq)))
    def peek(self):
        return max(self.pq)
    def size(self):
        return len(self.pq)
    def clear(self):
        self.pq=[]
    def getter(self):
        print(self.pq)
```

```python
p=Priority_Queue()
p.PriorityQueue()
print(p.size())
print(p.isEmpty())
p.enqueue([3,9])
p.enqueue([5,1])
p.enqueue([1,2])
print(p.peek())
```
